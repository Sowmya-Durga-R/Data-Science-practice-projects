# Apriori algorithm

```python
freq_itemset=apriori(book,min_support=0.1,use_colnames=True)
freq_itemset
```

```python
book[book["YouthBks"]==1].shape[0]/book.shape[0]
```

```python
rules=association_rules(freq_itemset,metric="lift",min_threshold=1.3)
rules
```

```python
rules=association_rules(freq_itemset,metric="lift",min_threshold=2)
rules
```

```python
rules[rules.lift>1]
```

```python
rules.sort_values('lift',ascending=False)[0:10]
```

```python
rules=association_rules(freq_itemset,metric="support",min_threshold=0.1)
rules
```

```python
rules.sort_values('support',ascending=False)[0:10]
```

```python
rules=association_rules(freq_itemset,metric="support",min_threshold=0.05)
rules
```

```python
rules.sort_values('support',ascending=True)[0:10]
```

```python
rules=association_rules(freq_itemset,metric="confidence",min_threshold=0.05)
rules
```

```python
rules.sort_values('confidence',ascending=False)[0:10]
```

```python
rules[rules.confidence>0.8]
```
